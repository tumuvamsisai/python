```python
#create box plots
fig,ax=plt.subplots(ncols=7,nrows=2,figsize=(20,10))
index=0
ax=ax.flatten()

for col,value in df.items():
        sns.boxplot(y=col,data=df, ax=ax[index])
        index+=1
plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)
```

```python
#create dist plots
fig,ax=plt.subplots(ncols=7,nrows=2,figsize=(20,10))
index=0
ax=ax.flatten()

for col,value in df.items():
        sns.distplot(value, ax=ax[index])
        index+=1
plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)
```

```python
df.drop(["Ash"],axis=1,inplace=True)
print(df)
X=df.drop(['Class'],axis=1)
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split

random_state = 2
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3,random_state=random_state,shuffle=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(X_train)
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
ypred=knn.predict(X_test)
acc_knn=accuracy_score(y_test,ypred)
print("accuracy",acc_knn)
```

```python
pred_test= knn.model_
```

```python
conclusion = pd.DataFrame({'model': ["KNN"],
                           'accuracy': accuracy_score(y_test,pred_knn)*100})
conclusion
```
