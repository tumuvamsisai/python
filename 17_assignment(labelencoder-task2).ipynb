```python
import numpy as np
import pandas as pd
dataset=pd.read_csv("ODI Rankings.csv")
df=pd.DataFrame(dataset)
df
```

```python
df=df.iloc[:,:].values
df
```

```python
#using labekencoder and coverting the categorical data to numerical for column "team"
from sklearn.preprocessing import LabelEncoder
country=LabelEncoder()
df[:,0]=country.fit_transform(df[:,0])
df


```

```python
#using onehotencoder to spilt the column into multiple columns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encode',OneHotEncoder(),[0])],remainder="passthrough")
```

```python
y=ct.fit_transform(df)
df=pd.DataFrame(y)
df
```
