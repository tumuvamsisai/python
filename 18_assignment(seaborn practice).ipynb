```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("ODI Rankings.csv")
df
```

```python
print(sns.get_dataset_names("iris"))
```

```python
df=sns.load_dataset("iris")
df
```

```python
#distribution plot
sns.distplot(df['petal_length'])
```

```python
#join plot
sns.jointplot(x="petal_length",y="petal_width",data=df,kind="hex")
```

```python
#kde plot
sns.kdeplot(df['sepal_length'])
```

```python
#pair plot
sns.pairplot(df)
```

```python
#rug plot
sns.rugplot(df['sepal_length'])
```

```python
pen_df=sns.load_dataset("penguins")
pen_df
```

```python
#barplot
sns.barplot(x="sex",y="body_mass_g",data=pen_df)
```

```python
#countplot
sns.countplot(x="sex",data=pen_df)
```

```python
#box plot
sns.boxplot(x="island",y="body_mass_g",data=pen_df,hue="sex")
```

```python
#violin plot
sns.violinplot(x="island",y="body_mass_g",data=pen_df,hue="sex")
```

```python
car_df=sns.load_dataset("car_crashes")
car_df
```

```python
#heatmap
plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale=1.4)
mx = car_df.corr()
sns.heatmap(mx,annot=True,cmap="Blues")
```

```python
#cluster map
flights=sns.load_dataset("flights")
flights=flights.pivot_table(index="month",columns="year",values="passengers")
sns.heatmap(flights,cmap="Blues",)
```

```python
#clustermap
iris=sns.load_dataset("iris")
species=iris.pop("species")
sns.clustermap(iris)
```

```python
#pairgrid
iris=sns.load_dataset("iris")
iris_g=sns.PairGrid(iris,hue="species")
iris_g.map(plt.scatter)
```
